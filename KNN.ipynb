```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.preprocessing  import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn import metrics
#get dataset list from the seaborn 
sns.get_dataset_names()
df=sns.load_dataset('tips')
df.head()
features=df.drop(['size'],axis=1)
y=df.filter(['size'],axis=1)
# having only nummerical dataset
numerical_dataset=features.drop(['sex','smoker','day','time'],axis=1)
print(numerical_dataset.head())
categorical_dataset=features.filter(['sex','smoker','day','time'])
print(categorical_dataset.head())
#perform one hot encoding 
encoding_dataset=pd.get_dummies(categorical_dataset,drop_first=True)
x=pd.concat([numerical_dataset,encoding_dataset],axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
#scaling
scaling_dataset=StandardScaler()
x_train=scaling_dataset.fit_transform(x_train)
x_test=scaling_dataset.transform(x_test)

#training the model using KNN(without cross validation)
knn_neighbor=KNeighborsRegressor(n_neighbors=10)

#training the model
regressor_model=knn_neighbor.fit(x_train,y_train)
#now make the prediction 
y_prediction=regressor_model.predict(x_test)

#Evaluating model
print("Mean absolute error;",metrics.mean_absolute_error(y_test,y_prediction))
print("Mean Square error;",metrics.mean_squared_error(y_test,y_prediction))
print("Root mean Squared error;",np.sqrt(metrics.mean_squared_error(y_test,y_prediction)))




```
